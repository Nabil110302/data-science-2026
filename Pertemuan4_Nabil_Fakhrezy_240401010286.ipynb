Pada praktikum ini, saya melakukan eksplorasi statistik sederhana menggunakan Python, Pandas, NumPy, Matplotlib, Seaborn, dan SciPy. Dataset yang digunakan adalah dataset Iris yang terdiri dari beberapa ukuran bunga seperti sepal length, sepal width, petal length, petal width, dan species.


## Langkah 1 - Load dan Inspect Dataset

Pada bagian ini, dataset Iris dimuat menggunakan `seaborn.load_dataset()`. Setelah itu, saya menampilkan ukuran dataset, tipe data, beberapa baris awal, dan statistik deskriptif awal.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

df = sns.load_dataset('iris')

print('Shape:', df.shape)
print('\nTipe data:')
print(df.dtypes)

print('\n5 data awal:')
display(df.head())

print('\nStatistik deskriptif:')
display(df.describe().round(3))
```

**Interpretasi:**

Dataset Iris memiliki **150 baris** dan **5 kolom**. Empat kolom berisi data numerik, yaitu `sepal_length`, `sepal_width`, `petal_length`, dan `petal_width`, sedangkan kolom `species` berisi kategori jenis bunga Iris.


## Langkah 2 - Statistik Deskriptif Lengkap

Pada bagian ini, saya menghitung beberapa ukuran statistik untuk setiap kolom numerik, yaitu mean, median, standar deviasi, varians, skewness, dan kurtosis.


```python
for col_name in df.select_dtypes(include='number').columns:
    col = df[col_name]
    print(f'\n=== {col_name} ===')
    print(f'Mean      : {col.mean():.3f}')
    print(f'Median    : {col.median():.3f}')
    print(f'Std Dev   : {col.std():.3f}')
    print(f'Varians   : {col.var():.3f}')
    print(f'Skewness  : {col.skew():.3f}')
    print(f'Kurtosis  : {col.kurt():.3f}')
```

**Interpretasi:**

Dari hasil statistik deskriptif, setiap kolom numerik memiliki karakteristik yang berbeda. Nilai mean dan median digunakan untuk melihat pemusatan data, sedangkan standar deviasi dan varians menunjukkan penyebaran data. Skewness digunakan untuk melihat kemiringan distribusi, sedangkan kurtosis digunakan untuk melihat bentuk ekor distribusi.


## Langkah 3 - Analisis Distribusi Sepal Length

Pada bagian ini, saya membuat histogram dan KDE untuk kolom `sepal_length`. Selain itu, saya menambahkan garis mean dan median untuk melihat posisi rata-rata dan nilai tengah data.


```python
fig, ax = plt.subplots(figsize=(8, 4))

sns.histplot(df['sepal_length'], kde=True, bins=20, ax=ax)

ax.axvline(
    df['sepal_length'].mean(),
    linestyle='--',
    label=f"Mean = {df['sepal_length'].mean():.2f}"
)

ax.axvline(
    df['sepal_length'].median(),
    linestyle='--',
    label=f"Median = {df['sepal_length'].median():.2f}"
)

ax.set_title('Distribusi Sepal Length')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Frekuensi')
ax.legend()

plt.show()

print('Mean sepal_length  :', round(df['sepal_length'].mean(), 3))
print('Median sepal_length:', round(df['sepal_length'].median(), 3))
print('Skewness           :', round(df['sepal_length'].skew(), 3))
```

**Interpretasi:**

Distribusi `sepal_length` terlihat cukup mendekati simetris karena nilai mean dan median tidak terlalu jauh. Skewness yang berada di sekitar nilai kecil menunjukkan bahwa distribusi tidak terlalu miring ke kiri atau kanan.


## Langkah 4 - Boxplot dan Violin Plot per Spesies

Pada bagian ini, saya membuat boxplot dan violin plot untuk melihat perbedaan `petal_length` berdasarkan spesies bunga Iris.


```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.boxplot(data=df, x='species', y='petal_length', ax=axes[0])
axes[0].set_title('Boxplot Petal Length per Spesies')
axes[0].set_xlabel('Species')
axes[0].set_ylabel('Petal Length')

sns.violinplot(data=df, x='species', y='petal_length', inner='box', ax=axes[1])
axes[1].set_title('Violin Plot Petal Length per Spesies')
axes[1].set_xlabel('Species')
axes[1].set_ylabel('Petal Length')

plt.tight_layout()
plt.show()

summary_species = df.groupby('species')['petal_length'].agg(['mean', 'median', 'min', 'max', 'std']).round(3)
display(summary_species)
```

**Interpretasi:**

Berdasarkan boxplot dan violin plot, spesies **setosa** memiliki nilai `petal_length` paling kecil. Spesies **virginica** cenderung memiliki `petal_length` paling besar. Sementara itu, penyebaran nilai pada spesies **versicolor** dan **virginica** terlihat lebih besar dibandingkan setosa.


## Langkah 5 - Matriks Korelasi Pearson

Pada bagian ini, saya menghitung matriks korelasi Pearson untuk melihat hubungan antar kolom numerik.


```python
corr = df.drop('species', axis=1).corr(method='pearson')

print('Matriks Korelasi Pearson:')
display(corr.round(3))

mask = np.triu(np.ones(corr.shape, dtype=bool))
corr_masked = corr.where(~mask)

max_pair = corr_masked.stack().idxmax()
max_value = corr.loc[max_pair]

min_pair = corr_masked.stack().idxmin()
min_value = corr.loc[min_pair]

print(f'Korelasi tertinggi: {max_pair} = {max_value:.3f}')
print(f'Korelasi terendah : {min_pair} = {min_value:.3f}')
```

**Interpretasi:**

Korelasi tertinggi terdapat pada pasangan **petal_length** dan **petal_width**. Artinya, semakin panjang petal bunga, umumnya semakin besar pula lebar petalnya. Korelasi terendah terdapat pada pasangan yang memiliki hubungan negatif, yaitu ketika peningkatan satu variabel cenderung diikuti penurunan variabel lainnya.


## Langkah 6 - Scatter Plot dan Heatmap Korelasi

Pada bagian ini, saya membuat scatter plot untuk pasangan variabel yang memiliki korelasi tinggi, yaitu `petal_length` dan `petal_width`. Selain itu, saya membuat heatmap untuk menampilkan matriks korelasi secara visual.


```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.scatterplot(
    data=df,
    x='petal_length',
    y='petal_width',
    hue='species',
    ax=axes[0]
)

sns.regplot(
    data=df,
    x='petal_length',
    y='petal_width',
    scatter=False,
    ax=axes[0]
)

axes[0].set_title('Scatter Plot: Petal Length vs Petal Width')
axes[0].set_xlabel('Petal Length')
axes[0].set_ylabel('Petal Width')

sns.heatmap(
    corr,
    annot=True,
    fmt='.2f',
    vmin=-1,
    vmax=1,
    square=True,
    ax=axes[1]
)

axes[1].set_title('Heatmap Korelasi Pearson')

plt.tight_layout()
plt.show()
```

**Interpretasi:**

Scatter plot menunjukkan hubungan positif yang kuat antara `petal_length` dan `petal_width`. Titik-titik data membentuk pola naik, sehingga dapat disimpulkan bahwa kedua variabel tersebut saling berkaitan cukup kuat. Heatmap juga memperkuat hasil ini karena nilai korelasi antara `petal_length` dan `petal_width` mendekati 1.
