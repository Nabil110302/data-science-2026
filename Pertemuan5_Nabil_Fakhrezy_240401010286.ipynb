## Dashboard Visualisasi Statis Dataset Tips

Pada tugas ini saya menggunakan dataset **Tips** dari Seaborn. Dataset ini berisi data tagihan pelanggan restoran, jumlah tip, jenis kelamin, status perokok, hari, waktu makan, dan jumlah orang dalam satu meja. Tujuan analisis ini adalah membuat dashboard visualisasi statis dan membaca insight dari grafik menggunakan pendekatan **What? So what? Now what?**


## 1. Import Library dan Load Dataset

Langkah pertama adalah mengimpor library yang dibutuhkan, kemudian memuat dataset `tips` dari Seaborn. Setelah itu, saya melihat ukuran dataset, tipe data, data awal, dan statistik deskriptif.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns

sns.set_theme(style='whitegrid', palette='Set2')

df = sns.load_dataset('tips')

print('Shape dataset:', df.shape)
print('\nTipe data:')
print(df.dtypes)

print('\n5 data awal:')
display(df.head())

print('\nStatistik deskriptif:')
display(df.describe().round(2))
```

**Interpretasi awal:**

Dataset Tips memiliki 244 baris dan 7 kolom. Kolom numerik yang dapat dianalisis adalah `total_bill`, `tip`, dan `size`, sedangkan kolom kategorik meliputi `sex`, `smoker`, `day`, dan `time`. Data ini cocok untuk visualisasi karena memiliki kombinasi variabel numerik dan kategorik.


## 2. Ringkasan Data

Sebelum membuat grafik, saya melihat rata-rata tagihan per hari dan rata-rata tip berdasarkan waktu makan.


```python
avg_bill_day = df.groupby('day', observed=False)['total_bill'].mean().round(2)
avg_tip_time = df.groupby('time', observed=False)['tip'].mean().round(2)
count_day = df['day'].value_counts()

print('Rata-rata total bill per hari:')
print(avg_bill_day)

print('\nRata-rata tip berdasarkan waktu makan:')
print(avg_tip_time)

print('\nJumlah transaksi per hari:')
print(count_day)
```

**Interpretasi ringkas:**

Ringkasan ini membantu memahami data sebelum divisualisasikan. Rata-rata `total_bill` per hari digunakan untuk melihat hari dengan tagihan tertinggi. Rata-rata `tip` berdasarkan waktu makan membantu melihat apakah pelanggan saat lunch atau dinner memberi tip yang berbeda.


## 3. Dashboard Visualisasi Statis

Dashboard ini terdiri dari empat grafik utama: bar chart rata-rata total tagihan per hari, histogram distribusi total tagihan berdasarkan waktu makan, boxplot total tagihan berdasarkan hari dan gender, serta scatter plot hubungan total tagihan dan tip.


```python
fig = plt.figure(figsize=(16, 12))
fig.suptitle('Dashboard Analisis Dataset Tips', fontsize=18, fontweight='bold', y=0.98)

gs = gridspec.GridSpec(2, 2, figure=fig, hspace=0.38, wspace=0.30)
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])

# Grafik 1: Bar Chart
avg_by_day = df.groupby('day', observed=False)['total_bill'].mean()
avg_by_day.plot(kind='bar', ax=ax1, edgecolor='white')
ax1.set_title('Rata-rata Total Tagihan per Hari')
ax1.set_xlabel('Hari')
ax1.set_ylabel('Total Bill (USD)')
ax1.tick_params(axis='x', rotation=0)
ax1.spines[['top', 'right']].set_visible(False)
for container in ax1.containers:
    ax1.bar_label(container, fmt='%.1f', padding=3, fontsize=9)

# Grafik 2: Histogram + KDE
sns.histplot(data=df, x='total_bill', hue='time', kde=True, alpha=0.6, ax=ax2)
ax2.set_title('Distribusi Total Tagihan berdasarkan Waktu Makan')
ax2.set_xlabel('Total Bill (USD)')
ax2.set_ylabel('Frekuensi')
ax2.spines[['top', 'right']].set_visible(False)

# Grafik 3: Boxplot
sns.boxplot(data=df, x='day', y='total_bill', hue='sex', ax=ax3)
ax3.set_title('Distribusi Total Tagihan per Hari dan Gender')
ax3.set_xlabel('Hari')
ax3.set_ylabel('Total Bill (USD)')
ax3.spines[['top', 'right']].set_visible(False)

# Grafik 4: Scatter Plot
sns.scatterplot(data=df, x='total_bill', y='tip', hue='time', size='size', sizes=(40, 220), alpha=0.75, ax=ax4)
sns.regplot(data=df, x='total_bill', y='tip', scatter=False, color='gray', ax=ax4)
ax4.set_title('Hubungan Total Tagihan dan Tip')
ax4.set_xlabel('Total Bill (USD)')
ax4.set_ylabel('Tip (USD)')
ax4.spines[['top', 'right']].set_visible(False)

plt.savefig('dashboard_tips_nabil.png', dpi=150, bbox_inches='tight')
plt.show()
print('Dashboard tersimpan sebagai dashboard_tips_nabil.png')
```

## 4. Insight Grafik 1 - Bar Chart Rata-rata Total Tagihan per Hari

**What?** Grafik bar chart menunjukkan rata-rata `total_bill` pada setiap hari dalam dataset Tips. Dari grafik, terlihat bahwa hari dengan rata-rata tagihan tertinggi berada pada akhir pekan, terutama Minggu, sedangkan hari lain cenderung lebih rendah.

**So what?** Hal ini menunjukkan bahwa pelanggan pada akhir pekan cenderung mengeluarkan uang lebih banyak ketika makan di restoran. Kemungkinan penyebabnya adalah pelanggan datang bersama keluarga atau teman, sehingga total tagihan menjadi lebih besar.

**Now what?** Restoran dapat lebih memperhatikan persediaan bahan makanan dan jumlah staf pada hari akhir pekan. Analisis lanjutan juga dapat dilakukan dengan melihat jumlah pelanggan (`size`) pada masing-masing hari.


## 5. Insight Grafik 2 - Histogram Total Tagihan berdasarkan Waktu Makan

**What?** Histogram menunjukkan sebaran nilai `total_bill` berdasarkan waktu makan, yaitu lunch dan dinner. Sebagian besar total tagihan berada pada rentang rendah sampai menengah, sementara hanya sebagian kecil pelanggan yang memiliki tagihan sangat tinggi.

**So what?** Distribusi ini menunjukkan bahwa kebanyakan pelanggan memiliki pola pengeluaran yang relatif normal, tetapi tetap ada beberapa transaksi bernilai tinggi. Tagihan pada waktu dinner terlihat lebih dominan karena jumlah observasi dinner lebih banyak dalam dataset.

**Now what?** Restoran dapat menggunakan informasi ini untuk memahami pola pembelian pelanggan saat lunch dan dinner. Jika ingin meningkatkan pendapatan pada lunch, restoran dapat membuat promo paket makan siang agar transaksi pada jam tersebut meningkat.


## 6. Insight Grafik 3 - Boxplot Total Tagihan per Hari dan Gender

**What?** Boxplot memperlihatkan distribusi `total_bill` berdasarkan hari dan gender. Grafik ini menunjukkan nilai median, rentang data, dan kemungkinan outlier pada setiap kelompok.

**So what?** Beberapa kelompok memiliki sebaran tagihan yang lebih lebar, artinya pengeluaran pelanggan pada kelompok tersebut lebih bervariasi. Outlier menunjukkan adanya pelanggan dengan tagihan yang jauh lebih tinggi dibandingkan pelanggan lain.

**Now what?** Restoran dapat menelusuri lebih lanjut kelompok pelanggan yang memiliki tagihan tinggi untuk memahami pola pesanan mereka. Informasi ini bisa digunakan untuk strategi promosi atau rekomendasi menu.


## 7. Insight Grafik 4 - Scatter Plot Total Tagihan dan Tip

**What?** Scatter plot menunjukkan hubungan antara `total_bill` dan `tip`. Secara umum, titik-titik membentuk pola naik, artinya semakin besar total tagihan, tip yang diberikan juga cenderung lebih besar.

**So what?** Pola ini menunjukkan adanya hubungan positif antara total tagihan dan tip. Pelanggan dengan total pembelian lebih besar umumnya memberikan tip lebih tinggi, meskipun besarannya tetap bervariasi.

**Now what?** Restoran dapat menjaga kualitas pelayanan terutama pada meja dengan jumlah pelanggan lebih banyak atau total tagihan yang tinggi. Analisis lanjutan dapat dilakukan untuk melihat apakah waktu makan atau jumlah pelanggan mempengaruhi besar tip.

Visualisasi data membantu memperjelas pola yang sulit dilihat jika hanya membaca tabel angka.
